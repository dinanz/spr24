# L1 Intro to databases
Chapter 1

### Database management system
- DBMS is a collection of data and the program to manage and access the data
- not the data itself
- the database is technically the data itself
- In general 2 modes in which databases are used
    1. Online transaction processing
    2. Data analytics
- CRUD
## Issues with storing information in files not database
- Data redundancy and inconsistency
- Difficulty of accessing the data
- Data isolation
- Integrity problems
- Atomicity problems
- Concurrent access anomalies
- Security problems
### Data models
- Relational model
    - Data seen as interrelated tables (relations)
    - Most popular model
- Entity-relationship model
    - Models relationships between "entities"
    - Maps onto relational model: table=entity; entity instance=tuple
    - Explicitly models cardinality
- Semi-structured data model
    - Different objects of same type may have different sets of attributes
    - Often implemented in nonrelational databases like MongoDB
- Object-based data model
    - Allows storage of objects from OOP languages in a column
### Data abstraction
- Physical level: how the data is stored
- Logical level: what data is stored, and relationships
- View level: simplifies user interactino with databse
### Database languages
- Data definition language (DDL)
    - Defines the database schema (how tables are related to one another)
    - In relational model:
        - Defines each table: column names and types
        - Defines constraints
- Data manipulation language (DML)
    - Defines commands/queries
    - CRUD
    - Two types of DMLs
        - Procedural: tells DBMS what info to retrive and method of retrieval
        - Declarative: tells DBMS what info to retrieve, but method is up to DBMS
            - The part of DML involved with info retrieval is the `query language`
            - SQL is a declarative DML
- DDL and DML make up a single language
### Accessing databases from application programs 
- Open database connectivity (ODBC), works with languages like C, Python, etc
- Java database connectivity (JDBC)
- Python has its own database API
### Database design
- Schema consists of design of each table
- Relationship between tables (1-1, 1-n, n-1, n-n)
- Design based on needs
### Database engine
- Heart of the system that stores the data and runs queries
1. Storage manager
    - Usually trees or hash tables
    - Components
        - `Authentication and integrity manager`
            - Enforces user authentication and privileges
            - Enforces integrity constraints like foreign key and uniqueness
        - Transaction
            - Ensures consistent state
            - Transactions must be complete, otherwise they will be rolled back
        - File manager
        - Buffer Manager
    - Data structures managed
        - Data files
        - Data dictionary
            - Metadata about databases
        - Indexes
2. Query processor
    - Translates queries into actual operations that retrieve data
    - Functions
        - DDL interpreter
            - Reads data definition statements and stores them into data dictionary
        - DML compiler translates DML statements into execution plan
            - Performs `query optimisation`, selects most efficient plan
        - Query execution engine
            - Executes lower level commands generated by DML compiler
3. Transaction manager
    - Ensure the following
        - Atomicity
        - Consistency
        - Durability
        - Must not violate consistency constraints
        - If inconsistent, roll back using `recovery manager`
        - Concurrency
### Database system architecture
- Two basic architectures
    1. two tier architecture: user interacts with application on their client machine, issueing queries to database on server machine
    2. three tier architecture, user interacts with application on client machine, sends messages to application server on server machine, then queries database system
### Database users and administrators
- 3 types of users
    1. Naive users
        - Like an online shopper, no direct knowledge or interactino with database
    2. Application programmers
    3. Sophisticated users
        - Do not wite programs but do write queries
- Database administrators (DBA)
    - Defines schemas, set up indices
    - Modify schemas
    - Grant permissions
    - Conduct maintenance
        - Monitor system jobs
        - Backups
        - Check storage space
