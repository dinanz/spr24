# The Relational Model

### Structure of a relational database
- Columns are attributes
- Rows are entries
- Primary key: usually one column, used to identify entries
- Attributes have constraints, or `domains`
- Relations/attributes are somewhat 
### Database schema
- The design of a table; table definition
- Relation schema: list of attributes foa a particular relation and their domains
- Relational instance: a relation with particular values; like a value of a variable
- A column can be primary key for one table and foreign key for another
### Key
- `Superkey`: set of attributes that uniquely identifies an entry
- `Candidate key`: a superkey that is minimal; no subset of it is also a superkey
#### Primary key
- May be more than 1 candidate key for a table
- `Primary key`: a candidate key that is chosen to be unique identifer for entries
    - Can be a single column or a composite primary key from multiple columns
    - Constraints: must be defined and unique
    - Defined in the CREATE
    - Often arbitrarily generated
- `Foreign keys`: reference to primary keys from another table
    - Defined during CREATE
### Query languages
- 3 types
    1. Imperative query langauge
        - Specifies sequence of operations on CRUD
    2. Functional query language
        - Expressed as series of functions to operate
    3. Declarative query language
        - Specifies information requested
        - e.g. SQL
